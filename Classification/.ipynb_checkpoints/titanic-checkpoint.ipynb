Missing data Imputation using MICE (Multivariate Imputation Chained Equations)

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
data['Sex']
```

```python
data
```

```python
missing_value = data[data['Embarked'].isna()].index.tolist()
missing_value
```

```python
women = data.loc[data['Sex'] == 0]['Survived']
ratio_women = sum(women) / len(women)
ratio_women
```

```python
man = data.loc[data['Sex'] == 1] ['Survived']
ratio_man = sum(man) / len(man)
ratio_man
```
