```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('laptop_pricing_dataset.csv')
data
```

```python
data = data.drop(['Unnamed: 0', 'Category'], axis=1)
data
```

```python
data = data.dropna()
data
```

```python
data.dtypes
```

```python
manufacturer = data['Manufacturer']
manufacturer
```

```python
manufacturer = pd.get_dummies(manufacturer)
manufacturer = manufacturer.astype(int)
manufacturer
```

```python
screen = data['Screen']
screen
```

```python
screen = pd.get_dummies(screen)
screen = screen.astype(int)
screen
```

```python
data = pd.concat([data, manufacturer, screen], axis=1)
```

```python
data
```

```python
data = data.drop(['Manufacturer', 'Screen'], axis=1)
data
```

```python
data.corr()
```

```python
data.hist(figsize=(12, 8))
```

```python
import seaborn as sns

plt.figure(figsize=(24,12))
sns.heatmap(data.corr(), annot=True, cmap="YlGnBu")   
```

```python
X = data.drop('Price', axis=1)
y = data['Price']
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.pipeline import Pipeline

pipeline_lr = Pipeline([
    ('scaler ', StandardScaler()),
    ('regression ', LinearRegression())
])

pipeline_lr.fit(X_train, y_train)
```

```python
y_lr_pred = pipeline_lr.predict(X_test)
```

```python

pipeline_rf = Pipeline([
    ('regression ', RandomForestRegressor())
])
pipeline_rf.fit(X_train, y_train)
```

```python
y_rf_pred = pipeline_rf.predict(X_test)
```

```python
out = np.c_[y_test, y_lr_pred, np.abs((y_test - y_lr_pred)/y_test), y_rf_pred, np.abs((y_test - y_rf_pred)/y_test)]
out = pd.DataFrame(out, columns = ['y_test',
                                   'y_lr_pred', 'linear_regression',
                                   'y_rf_pred', 'random_forest'])
melt_out = out[['linear_regression', 'random_forest']].melt()
sns.displot(data = melt_out, x='value', hue='variable', kind='kde')
```
