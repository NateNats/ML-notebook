```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('apple_quality.csv')
data
```

```python
data = data.dropna()
```

```python
data.head()
```

```python
compare = data['Quality'].value_counts()
data.drop('A_id', axis=1, inplace=True)
```

```python
data
```

```python
plt.bar(compare.index, compare, color='blue')
plt.xlabel('quality')
plt.ylabel('count')
plt.title('Apple Quality')
plt.show()
```

```python
X = data.drop('Quality', axis=1)
features = ['Size', 'Weight', 'Sweetness', 'Crunchiness', 'Juiciness', 'Ripeness', 'Acidity']

for col in features:
    plt.scatter(data[col], data['Size'])
    plt.xlabel(col)
    plt.ylabel('values')
    plt.title(col)
    plt.show()
```

```python
for col in features:
    plt.boxplot(data[col])
    plt.xlabel(col)
    plt.ylabel('values')
    plt.title(col)
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y = le.fit_transform(y)
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(criterion='entropy', random_state=10 ,max_depth=12)
classifier.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, mean_squared_error, r2_score

y_train_pred = classifier.predict(X_train)
y_test_pred = classifier.predict(X_test)
print('Model Accuracy: {:.2f}%'.format(accuracy_score(y_train, y_train_pred)))
print('Test Accuracy: {:.2f}%'.format(accuracy_score(y_test, y_test_pred)))
```

```python
mse = mean_squared_error(y_test, y_test_pred)
print('Mean Squared Error: {:.4f}%'.format(mse))
```

```python
r2_score = r2_score(y_test, y_test_pred)
print('R2 Score: {:.4f}'.format(r2_score))
```

```python
from sklearn.neighbors import KNeighborsClassifier

classifier = KNeighborsClassifier(n_neighbors= 10)
classifier.fit(X_train, y_train)
classifier
```

```python
y_train_pred = classifier.predict(X_train)
y_test_pred = classifier.predict(X_test)
print('Model Accuracy: {:.2f}%'.format(accuracy_score(y_train, y_train_pred)))
print('Test Accuracy: {:.2f}%'.format(accuracy_score(y_test, y_test_pred)))
```

```python
mse = mean_squared_error(y_test, y_test_pred)
print('Mean Squared Error test: {:.2f}'.format(mse))
```

```python
mse = mean_squared_error(y_train, y_train_pred)
print('Mean Squared Error Model: {:.2f}'.format(mse))
```

```python
from sklearn.metrics import classification_report
report = classification_report(y_test, y_test_pred)
print(report)
```

```python
report = classification_report(y_train, y_train_pred)
print(report)
```
