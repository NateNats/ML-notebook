```python
data['Income'] = data['Income'].fillna(data['Income'].mean())
data
```

```python
from datetime import date

data['Age'] = date.today().year - data['Year_Birth']
data
```

```python
data['Year'] = pd.DatetimeIndex(data['Dt_Customer']).year
data['Enrollment_Length'] = date.today().year - data['Year']
data
```

```python
data.columns
```

```python
data.dtypes
```

```python
data['Income$'] = data['Income'].astype(str).str.replace(',', '').str.replace('$', '').str.replace('.', '').fillna(0).astype(np.int64)
data['Income_M$'] = data['Income$'].apply(lambda x: round(x/1000000))
print(data[['Income_M$', 'Income$']])
```

```python
import matplotlib.pyplot as plt

log_var = ['Income_M$', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases', 'NumWebPurchases', 'NumCatalogPurchases']

fig = plt.figure(figsize=(24,10))

for j in range(len(log_var)):
    var = log_var[j]
    transformed = 'log_' + var
    data[transformed] = np.log10(data[var]+1)
    
    sub = fig.add_subplot(2, 5, j+1)
    sub.set_xlabel(var)
    data[transformed].plot(kind = 'hist')
```

```python
data
```

```python
clip_var = ['Age', 'NumWebVisitsMonth']

for i in clip_var:
    transformed = 'clipped_' + i
    
    upper_limit = data[i].quantile(0.95)
    lower_limit = data[i].quantile(0.05)
    data[transformed] = data[i].clip(lower_limit, upper_limit, axis=0)
    
    print(data[i].describe())
    print(data[transformed].describe())
```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler

scalers_list = [StandardScaler(), MinMaxScaler(), RobustScaler()]
scale_var = ['Enrollment_Length', 'Recency', 'NumStorePurchases', 'clipped_NumWebVisitsMonth', 'clipped_Age']

for i in range(len(scalers_list)):
    scaler = scalers_list[i]
    fig = plt.figure(figsize=(26, 5))
    plt.title(scaler, fontsize=20)
    
    for j in range(len(scale_var)):
        var = scale_var[j]
        scaled_var = 'scaled_' + var
        data[scaled_var] = scaler.fit_transform(data[var].values.reshape(-1,1))
        
        sub = fig.add_subplot(1, 5, j + 1)
        sub.set_xlabel(var)
        data[scaled_var].plot(kind = 'hist')
```

```python
scale_var = ['Enrollment_Length', 'Recency', 'NumStorePurchases', 'clipped_NumWebVisitsMonth', 'clipped_Age']

for i in range(len(scale_var)):
    fig = plt.figure(figsize=(26, 5))
    
    sub = fig.add_subplot(1, 5, i + 1)
    sub.set_xlabel(scale_var[i])
    data[scale_var[i]].plot(kind = 'hist')
```
